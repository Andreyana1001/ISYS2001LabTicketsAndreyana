## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
print('''My Budget Tracker Prompt:
write a python program that:

Allows users to input their monthly income
Allows users to input up to 3 expenses
Calculates remaining budget
Shows a warning if budget is negative''')
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# # Ask for monthly income
income = float(input("Enter your monthly income: "))

# Ask for up to 3 expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n--- Budget Summary ---")
print(f"Income: ${income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

# Show warning if budget is negative
if remaining_budget < 0:
    print("⚠️ Warning: Your budget is negative! You are overspending.")
else:
    print("✅ You are within your budget. Keep it up!")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: 📝 Detailed budget summary with advice"
enhancement_prompt = """
create a python program for a detailed budget summary with advice
"""

print(f"🚀 ENHANCEMENT: 📝 Detailed budget summary with advice")
```

```python
# ENHANCED BUDGET TRACKER
# # Detailed Budget Summary Program with Advice

# Get monthly income
income = float(input("Enter your monthly income: "))

# Get expenses
rent = float(input("Enter rent expense: "))
food = float(input("Enter food expense: "))
transport = float(input("Enter transport expense: "))

# Calculate totals
total_expenses = rent + food + transport
remaining_budget = income - total_expenses

# Display budget summary
print("\n===== BUDGET SUMMARY =====")
print(f"Monthly Income: ${income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

# Give advice based on budget status
print("\n===== BUDGET ADVICE =====")
if remaining_budget < 0:
    print("⚠️ You are spending more than you earn.")
    print("Advice: Try reducing non-essential expenses and review your budget.")
elif remaining_budget == 0:
    print("⚠️ Your budget is exactly balanced.")
    print("Advice: Aim to save a small amount for emergencies.")
else:
    print("✅ Great job! You have money left after expenses.")
    print("Advice: Consider saving or investing your remaining budget.")

print("\nThank you for using the budget planner 😊")
```
